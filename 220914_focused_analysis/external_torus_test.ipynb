```python
r1, r2, r3, r4 = symbols('r1 r2 r3 r4')
s1, s2, s3, s4 = symbols('s1 s2 s3 s4')
a11, a12, a21, a22 = symbols('a11 a12 a21 a22')
b11, b12, b21, b22 = symbols('b11 b12 b21 b22')
a13, a14, a23, a24 = symbols('a13 a14 a23 a24')
b13, b14, b23, b24 = symbols('b13 b14 b23 b24')
omega1, omega2 = symbols('omega1 omega2')
epsilon = symbols('epsilon')
x, y = symbols('x y')

p1 = a11 * sin(2 * pi * (r1 * x + s1 * y + b11)) + a12 * sin(2 * pi * (r2 * x + s2 * y + b12)) + a13 * sin(2 * pi * (r3 * x + s3 * y + b13)) + a14 * sin(2 * pi * (r4 * x + s4 * y + b14))
p2 = a21 * sin(2 * pi * (r1 * x + s1 * y + b21)) + a22 * sin(2 * pi * (r2 * x + s2 * y + b22)) + a23 * sin(2 * pi * (r3 * x + s3 * y + b23)) + a24 * sin(2 * pi * (r4 * x + s4 * y + b24))

t1 = x + omega1 + (epsilon / (2 * pi)) * p1
t2 = y + omega2 + (epsilon / (2 * pi)) * p2
```

```python
@njit
def p1(x, y):
    return (
        -0.2680 * np.sin(2*np.pi*(1 * x + 0 * y + 0.9850)) +
        -0.9106 * np.sin(2*np.pi*(0 * x + 1 * y + 0.5040)) +
        +0.3000 * np.sin(2*np.pi*(1 * x + 1 * y + 0.9470)) +
        -0.0400 * np.sin(2*np.pi*(0 * x - 1 * y + 0.2334))
    )

@njit
def p2(x, y):
    return (
        +0.0800 * np.sin(2*np.pi*(1 * x + 0 * y + 0.990)) +
        -0.5600 * np.sin(2*np.pi*(0 * x + 1 * y + 0.330)) +
        +0.9470 * np.sin(2*np.pi*(1 * x + 1 * y + 0.290)) +
        -0.4003 * np.sin(2*np.pi*(0 * x - 1 * y + 0.155))
    )

@njit
def torus_step(x, y):
    x_next = np.mod(x + 0.71151134457776362264681206697006238 + (0.4234823 / (2 * np.pi)) * p1(x, y), 1)
    y_next = np.mod(y + 0.87735009811261456100917086672849971 + (0.4234823 / (2 * np.pi)) * p2(x, y), 1)
    return x_next, y_next
```

```python
epsilon = 0.4234823
omega1 = 0.71151134457776362264681206697006238
omega2 = 0.87735009811261456100917086672849971
a11 = -0.2680
a12 = -0.9106
a13 = +0.3000
a14 = -0.0400
a21 = +0.0800
a22 = -0.5600
a23 = +0.9470
a24 = -0.4003
b11 = 0.9850
b12 = 0.5040
b13 = 0.9470
b14 = 0.2334
b21 = 0.9900
b22 = 0.3300
b23 = 0.2900
b24 = 0.1550
r1 = 1
r2 = 0
r3 = 1
r4 = 0
s1 = 0
s2 = 1
s3 = 1
s4 = -1


def tangent_map(x, y):
    return np.array([
    [
        epsilon*(2*np.pi*a11*r1*np.cos(np.pi*(2*b11 + 2*r1*x + 2*s1*y)) + 2*np.pi*a12*r2*np.cos(np.pi*(2*b12 + 2*r2*x + 2*s2*y)) + 2*np.pi*a13*r3*np.cos(np.pi*(2*b13 + 2*r3*x + 2*s3*y)) + 2*np.pi*a14*r4*np.cos(np.pi*(2*b14 + 2*r4*x + 2*s4*y)))/(2*np.pi) + 1,
        epsilon*(2*np.pi*a11*s1*np.cos(np.pi*(2*b11 + 2*r1*x + 2*s1*y)) + 2*np.pi*a12*s2*np.cos(np.pi*(2*b12 + 2*r2*x + 2*s2*y)) + 2*np.pi*a13*s3*np.cos(np.pi*(2*b13 + 2*r3*x + 2*s3*y)) + 2*np.pi*a14*s4*np.cos(np.pi*(2*b14 + 2*r4*x + 2*s4*y)))/(2*np.pi)
    ],
    [
        epsilon*(2*np.pi*a21*r1*np.cos(np.pi*(2*b21 + 2*r1*x + 2*s1*y)) + 2*np.pi*a22*r2*np.cos(np.pi*(2*b22 + 2*r2*x + 2*s2*y)) + 2*np.pi*a23*r3*np.cos(np.pi*(2*b23 + 2*r3*x + 2*s3*y)) + 2*np.pi*a24*r4*np.cos(np.pi*(2*b24 + 2*r4*x + 2*s4*y)))/(2*np.pi),
        epsilon*(2*np.pi*a21*s1*np.cos(np.pi*(2*b21 + 2*r1*x + 2*s1*y)) + 2*np.pi*a22*s2*np.cos(np.pi*(2*b22 + 2*r2*x + 2*s2*y)) + 2*np.pi*a23*s3*np.cos(np.pi*(2*b23 + 2*r3*x + 2*s3*y)) + 2*np.pi*a24*s4*np.cos(np.pi*(2*b24 + 2*r4*x + 2*s4*y)))/(2*np.pi) + 1
    ]])

@njit
def matrix_vector(x, y, x1, y1):
    return ( 
        (epsilon*(2*np.pi*a11*r1*np.cos(np.pi*(2*b11 + 2*r1*x + 2*s1*y)) + 2*np.pi*a12*r2*np.cos(np.pi*(2*b12 + 2*r2*x + 2*s2*y)) + 2*np.pi*a13*r3*np.cos(np.pi*(2*b13 + 2*r3*x + 2*s3*y)) + 2*np.pi*a14*r4*np.cos(np.pi*(2*b14 + 2*r4*x + 2*s4*y)))/(2*np.pi) + 1) * x1 + (epsilon*(2*np.pi*a11*s1*np.cos(np.pi*(2*b11 + 2*r1*x + 2*s1*y)) + 2*np.pi*a12*s2*np.cos(np.pi*(2*b12 + 2*r2*x + 2*s2*y)) + 2*np.pi*a13*s3*np.cos(np.pi*(2*b13 + 2*r3*x + 2*s3*y)) + 2*np.pi*a14*s4*np.cos(np.pi*(2*b14 + 2*r4*x + 2*s4*y)))/(2*np.pi)) * y1,
        (epsilon*(2*np.pi*a21*r1*np.cos(np.pi*(2*b21 + 2*r1*x + 2*s1*y)) + 2*np.pi*a22*r2*np.cos(np.pi*(2*b22 + 2*r2*x + 2*s2*y)) + 2*np.pi*a23*r3*np.cos(np.pi*(2*b23 + 2*r3*x + 2*s3*y)) + 2*np.pi*a24*r4*np.cos(np.pi*(2*b24 + 2*r4*x + 2*s4*y)))/(2*np.pi)) * x1 + (epsilon*(2*np.pi*a21*s1*np.cos(np.pi*(2*b21 + 2*r1*x + 2*s1*y)) + 2*np.pi*a22*s2*np.cos(np.pi*(2*b22 + 2*r2*x + 2*s2*y)) + 2*np.pi*a23*s3*np.cos(np.pi*(2*b23 + 2*r3*x + 2*s3*y)) + 2*np.pi*a24*s4*np.cos(np.pi*(2*b24 + 2*r4*x + 2*s4*y)))/(2*np.pi) + 1) * y1
    )
    
```

```python
steps = int(1e7)
x0 = 0.5
y0 = 0.5
x = np.zeros(steps)
y = np.zeros(steps)
x[0] = x0
y[0] = y0
disp_values = np.zeros(steps)
disp_vector_x = 1.0
disp_vector_y = 0.0
for i in tqdm(range(1, steps)):
    # tangent = tangent_map(x[i], y[i])
    disp_vector_x, disp_vector_y = matrix_vector(x[i-1], y[i-1], disp_vector_x, disp_vector_y)
    disp_values[i] = np.log(np.sqrt(disp_vector_x**2 + disp_vector_y**2))
    norm = np.sqrt(disp_vector_x**2 + disp_vector_y**2)
    disp_vector_x /= norm
    disp_vector_y /= norm
    x[i], y[i] = torus_step(x[i - 1], y[i - 1])
```

```python
@njit
def birkhoff_weights(n):
    weights = np.arange(n, dtype=np.float64)
    weights /= (n-1)
    weights = np.exp(-1 / (weights * (1 - weights)))
    return weights / np.sum(weights)
```

```python
n_vals = np.arange(10000, len(disp_values), 10000)
raw_average = np.zeros(len(n_vals))
birk_average = np.zeros(len(n_vals))

for idx, i in tqdm(enumerate(n_vals), total=len(n_vals)):
    raw_average[idx] = (np.mean(disp_values[0:i]))
    birk_average[idx] = (np.sum(disp_values[0:i] * birkhoff_weights(i)))
```

```python
plt.plot(n_vals, raw_average/n_vals, label='raw')
plt.plot(n_vals, birk_average/n_vals, label='birkhoff')
plt.legend()
plt.xscale("log")
plt.yscale("log")
```
